```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
df = pd.read_csv("Grocery Products Purchase.csv")
df.head()
```

```python
df
```

```python
len(df)
```

```python
records = []
for i in range(0, 9835):
    records.append([str(df.values[i,j]) for j in range(0, 32)])
```

```python
association_rule = apriori(transactions=records, min_support=0.50, min_confidence=0.4, min_lift=1.2, min_length=2)
association_results = list(association_rule)
```

```python
print(len(association_results))
```
